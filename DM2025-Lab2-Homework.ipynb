## 1. Model Development (10 pts Required)
### 1.1 Preprocessing Steps

I focused the preprocessing phase on preparing noisy, user-generated text for downstream modeling. Because the dataset originated from informal social media posts, it contained emojis, inconsistent punctuation, hashtags, and colloquial expressions. For transformer-based models, I applied minimal preprocessing to preserve these emotionally relevant cues. I only removed URLs, malformed unicode, and redundant whitespace. This approach aligns with modern research indicating that large language models rely on full textual context, including punctuation and emotive symbols, to capture sentiment. I also constructed a secondary "cleaned" version of the text for classical ML models involving lowercasing, whitespace normalization, and the removal of extraneous characters.

I reconstructed the dataset by combining raw text from JSON files, split identifiers, and emotion labels. Once these were merged into training and testing sets, we analyzed the distribution and found a significant class imbalance. The "joy" category was dominant, containing nearly twenty times as many examples as the "disgust" category. This disparity proved to be a central factor in the model's performance later on.

### 1.2 Feature Engineering Steps

I adapted my feature engineering techniques to fit the specific modeling strategy. For the classical machine-learning models, I employed TF-IDF vectorization to convert text into numerical representations. I used unigrams and bigrams to capture both individual tokens and short contextual patterns. Additionally, I used sublinear term weighting and a high-capacity vocabulary to retain expressive detail. These engineered features allowed linear models like Support Vector Machines to establish a strong baseline.

In contrast, for the transformer-based architectures (DistilBERT, RoBERTa, and DeBERTa), I relied primarily on the learned contextual embeddings produced by their own tokenizers. I added no handcrafted features to these models because the deep contextual representations generated by their attention mechanisms serve as effective end-to-end feature extractors.

### 1.3 Explanation of Your Model

I selected DeBERTa-v3-Large as the final model for this project. Developed by Microsoft, this state-of-the-art architecture improves upon previous models like BERT and RoBERTa by separating content and positional embeddings. These design choices allow the model to better understand subtle emotional signals, including punctuation patterns, emojis, and the informal phrasing common in social media text. Given that emotion detection requires interpreting these nuanced cues, I found that DeBERTa-v3-Large provided a significantly more capable representation than smaller architectures.

Before fine-tuning, I addressed the severe class imbalance in the dataset which initially limited performance for minority emotions like disgust and fear. I oversampled the training data to ensure that each emotion class had an approximately equal number of examples. After balancing the dataset, I fine-tuned the model on the emotion classification task. The combination of the model's large capacity and the balanced data allowed it to learn richer patterns and achieve stronger generalization across all emotion classes.

```python

import pandas as pd
import json

# Load raw JSON with posts
with open("final_posts.json", "r") as f:
    posts = json.load(f)

posts_df = pd.DataFrame(posts)[["id", "text"]]

# Load train/test split info
split_df = pd.read_csv("data_identification.csv")

# Load emotion labels
emotion_df = pd.read_csv("emotion.csv")

# Merge posts with split info
merged = posts_df.merge(split_df, on="id", how="left")

# Split into train and test
train_df = merged[merged["split"] == "train"].merge(emotion_df, on="id", how="left")
test_df  = merged[merged["split"] == "test"]

train_df = train_df[["id", "text", "emotion"]].reset_index(drop=True)
test_df  = test_df[["id", "text"]].reset_index(drop=True)
```
